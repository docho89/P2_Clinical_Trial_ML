```python
#Access the Components: The components_ attribute contains the principal axes in feature space.

loadings = pca.components_
#Identify Contributions: For the first principal component, you can look at the first row of the loadings array.

first_pc_loadings = loadings[0]
#Find the Most Contributing Features: You can then sort the absolute values of these loadings to identify which features contribute the most.

import numpy as np

# Create a DataFrame for better visualization
loading_df = pd.DataFrame({
    'Feature': df_cluster_scale.columns,
    'Loading': first_pc_loadings
})

# Sort by absolute loading values
loading_df['Absolute Loading'] = np.abs(loading_df['Loading'])
sorted_loading_df = loading_df.sort_values(by='Absolute Loading', ascending=False)

print(sorted_loading_df)
```

```python
# Create a a list to store inertia values and the values of k
inertia = []
k = list(range(1, 11))

# Create a for-loop where each value of k is evaluated using the K-means algorithm
# Fit the model using the service_ratings DataFrame
# Append the value of the computed inertia from the `inertia_` attribute of the KMeans model instance
for i in k:
    k_model = KMeans(n_clusters=i, n_init='auto', random_state=0)
    k_model.fit(clinical_pca_df)
    inertia.append(k_model.inertia_)

# Define a DataFrame to hold the values for k and the corresponding inertia
elbow_data = {"k": k, "inertia": inertia}

# Create the DataFrame from the elbow data
df_elbow = pd.DataFrame(elbow_data)

# Review the DataFrame
df_elbow.head()
```

```python
# Plot the DataFrame
df_elbow.plot.line(
    x="k", 
    y="inertia", 
    title="Elbow Curve", 
    xticks=k)
```

# K-nearest neighbors

```python
# Loop through different k values to find which has the highest accuracy.
# Note: We use only odd numbers because we don't want any ties.

from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
train_scores = []
test_scores = []
for k in range(1, 20, 2):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train_scaled, y_train)
    train_score = knn.score(X_train_scaled, y_train)
    test_score = knn.score(X_test_scaled, y_test)
    train_scores.append(train_score)
    test_scores.append(test_score)
    print(f"k: {k}, Train/Test KNN Score: {train_score:.3f}/{test_score:.3f}")
    
# Plot the results
plt.plot(range(1, 20, 2), train_scores, marker='o', label="training scores")
plt.plot(range(1, 20, 2), test_scores, marker="x", label="testing scores")
plt.xlabel("k neighbors")
plt.ylabel("accuracy score")
plt.legend()
plt.show()
```

```python
# Choose the best k, and refit the KNN classifier by using that k value.
# Note that k: 9 provides the best accuracy where the classifier starts to stablize
knn = KNeighborsClassifier(n_neighbors=13)
knn.fit(X_train_scaled, y_train)

# Print the score for the test data.
print('k=13 Test Acc: %.3f' % knn.score(X_test_scaled, y_test))
```

```python
# optimazation 
# Create a loop to vary the max_depth parameter
# Make sure to record the train and test scores 
# for each pass.

# Depths should span from 1 up to 40 in steps of 2
depths = range(1, 40, 2)

# The scores dataframe will hold depths and scores
# to make plotting easy
scores = {'train': [], 'test': [], 'depth': []}

# Loop through each depth (this will take time to run)
for depth in depths:
    rfc = RandomForestClassifier(max_depth=depth)
    rfc.fit(X_train_scaled, y_train)

    train_score = rfc.score(X_train_scaled, y_train)
    test_score = rfc.score(X_test_scaled, y_test)

    scores['depth'].append(depth)
    scores['train'].append(train_score)
    scores['test'].append(test_score)

# Create a dataframe from the scores dictionary and
# set the index to depth
scores_df = pd.DataFrame(scores).set_index('depth')

scores_df
```

```python

from sklearn.pipeline import Pipeline
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Lasso   #regularized linear regression
from sklearn.linear_model import Ridge   #regularized linear regression

models = {
    "SVC": SVC(),
    "RFC": RandomForestClassifier(),
    "HGBC": GradientBoostingClassifier()
}

pipes = {}

for acronym, model in models.items():
    pipes[acronym] = Pipeline([('model', model)])

param_grids = {}

# Parameter grid for GRADIENT BOOSTING
learning_rate_grids = [10 ** i for i in range(-4, 2)]
min_samples_leaf_grids = [1, 20, 100]

# Update param_grids
param_grids['HGBC'] = {
    'model__learning_rate': learning_rate_grids,
    'model__min_samples_leaf': min_samples_leaf_grids
}

# Example parameter grids for other models (you can adjust these as needed)
param_grids['SVC'] = {
    'model__C': [0.1, 1, 10],
    'model__gamma': ['scale', 'auto']
}

param_grids['RFC'] = {
    'model__n_estimators': [50, 100, 200],
    'model__max_depth': [None, 10, 20]
   
}


# The list of [best_score_, best_params_, best_estimator_] obtained by GridSearchCV
best_score_params_estimator_gs = []

# For each model
for acronym in pipes.keys():
    print("===============================")
    print(f"Optimizing {acronym} .........")
    # GridSearchCV
    gs = GridSearchCV(estimator=pipes[acronym],
                      param_grid=param_grids[acronym],
                      scoring='accuracy',  # Change this to an appropriate scoring metric
                      n_jobs=-1,
                      cv=5,
                      return_train_score=True,
                      verbose=1)
    
    # Fit the pipeline
    gs.fit(X_train, y_train)
    
    # Update best_score_params_estimator_gs
    best_score_params_estimator_gs.append([acronym, gs.best_score_, gs.best_params_, gs.best_estimator_])
    
    print(f"{acronym} best accuracy score: {gs.best_score_:.4f}")

# Sort by best score
best_score_params_estimator_gs.sort(key=lambda x: x[1], reverse=True)  # Sort by best score

results_df = pd.DataFrame(
    best_score_params_estimator_gs,
    columns=['model name', 'best_score', 'best_params', 'best_estimator']
)

results_df.sort_values(by='best_score', ascending=False, inplace=True)
results_df
```
